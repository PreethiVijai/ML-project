```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import data
```

```python
data.education["State"].value_counts().plot.bar(title="Counties by state", figsize=(12,4))
```

```python
supplemental = data.food["SUPPLEMENTAL"]
population = supplemental["Population Estimate, 2016"]
population.mean(), population.median(), population.std()
```

```python
bachelors = data.education["Percent of adults with a bachelor's degree or higher, 2014-18"]

bachelors.plot.hist(bins=range(0,100,5))
plt.title("Percent of adults with a bachelor's degree or higher, 2014-18")
plt.xticks(ticks=range(0,100,5), labels=range(0,100,5))
plt.ylabel("Number of counties")
plt.xlabel("Percent")
plt.show()
```

```python
print("Median:", bachelors.median())
print("Mean:", bachelors.mean(), "(", bachelors.std(), ")")
```
